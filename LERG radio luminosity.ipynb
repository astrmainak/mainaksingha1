```python
hdu = fits.open('/Volumes/Seagate/LERG/Radio Loud Table.fits')
data = hdu[1].data
z = data['z_1']
radio_class = data['RCl']
L = data['L']
```

```python
radio_flux = data['SNVSS']
hdu.close()
cosmo = FlatLambdaCDM(H0=70 * u.km / u.s / u.Mpc, Om0=0.3)
D_L = cosmo.luminosity_distance(z)
dist = D_L.to(u.cm)
log_radio_power = np.log10(4*np.pi) + 2*np.log10(dist.value) + np.log10(radio_flux) - 33
```

```python
select_compact_nearby = (L==1) & (radio_class==1) & (z<0.05)
select_extended_nearby = (L==1) & (radio_class==2) & (z<0.05)
select_compact_intermediate = (L==1) & (radio_class==1) & (z>0.05) & (z<0.1)
select_extended_intermediate = (L==1) & (radio_class==2) & (z>0.05) & (z<0.1)
select_compact_far = (L==1) & (radio_class==1) & (z>0.1) & (z<0.15)
select_extended_far = (L==1) & (radio_class==2) & (z>0.1) & (z<0.15)

log_P_rad_compact_nearby = log_radio_power[select_compact_nearby]
log_P_rad_extended_nearby = log_radio_power[select_extended_nearby]
log_P_rad_compact_intermediate = log_radio_power[select_compact_intermediate]
log_P_rad_extended_intermediate = log_radio_power[select_extended_intermediate]
log_P_rad_compact_far = log_radio_power[select_compact_far]
log_P_rad_extended_far = log_radio_power[select_extended_far]
```

```python
fig = plt.figure(figsize=(9,9))
plt.hist(log_P_rad_compact_nearby,bins=50,color='darkslategrey',alpha=0.4,label='FIRST 1 component (z < 0.05)')
plt.hist(log_P_rad_extended_nearby,bins=50,color='brown',alpha=0.6,label='FIRST 2 components (z < 0.05)')
plt.xlabel(r'log(P$_{1.4\ GHz}$)',fontsize=18)
plt.ylabel(r'Number',fontsize=18)
plt.legend(borderpad=1,fontsize=14)
```
